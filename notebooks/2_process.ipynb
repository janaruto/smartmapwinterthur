```python
import numpy as np

# Columns with 1–7 scale
likert_cols = [
    'Freie Marktwirtschaft',
    'Staatliche Umverteilung',
    'Bestrafung von Kriminellen',
    'Umweltschutz'
]

# Initialize scores
df['Links_Rechts'] = 0.0
df['Liberal_Konservativ'] = 0.0

# Normal answer mapping
answer_weights = {
    'Yes': 1,
    'Rather yes': 0.5,
    'Rather no': -0.5,
    'No': -1
}

# Helper to convert answer → numeric weight
def get_weight(val, col):
    if pd.isna(val):
        return 0

    # Special Likert columns
    if col in likert_cols:
        try:
            return (float(val) - 4) / 3 * 3
        except:
            return 0

    # Normal columns
    return answer_weights.get(val, 0) 


# Main scoring loop
for col, orientations in smartvote_mapping_winterthur_2026.items():

    if col not in df.columns:
        continue

    lr_orientation = orientations['Links_Rechts']
    lk_orientation = orientations['Liberal_Konservativ']

    # Convert column to numeric weights
    w = df[col].apply(lambda v: get_weight(v, col))

    # Links-Rechts axis
    if lr_orientation == 'Rechts':
        df['Links_Rechts'] += w
    elif lr_orientation == 'Links':
        df['Links_Rechts'] -= w

    # Liberal-Konservativ axis
    if lk_orientation == 'Konservativ':
        df['Liberal_Konservativ'] += w
    elif lk_orientation == 'Liberal':
        df['Liberal_Konservativ'] -= w


# Optional rounding
df['Links_Rechts'] = df['Links_Rechts'].round(2)
df['Liberal_Konservativ'] = df['Liberal_Konservativ'].round(2)

print(df[['Name', 'Party', 'Links_Rechts', 'Liberal_Konservativ']])
```
